```python
import pandas as pd
from sklearn.model_selection import StratifiedKFold, cross_val_score
from lightgbm import LGBMClassifier
from sklearn.preprocessing import LabelEncoder

# Load embeddings CSV
df = pd.read_csv("embedded/FAAMG_emb.csv")

# Prepare features and labels
label = 'label'
X = df.drop(columns=['id', label])
y = LabelEncoder().fit_transform(df[label])
print("Num of labels", len(df[label].unique()))

# Configure LGBMClassifier
clf = LGBMClassifier(
    objective="multiclass",
    num_class=len(df[label].unique()),
    learning_rate=0.01,
    num_leaves=50,
    # device='gpu',
    n_estimators=500,
    random_state=42,
    verbose=-1
)

# Set StratifiedKFold
cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=88)

# Cross-Validation
scores = cross_val_score(clf, X, y, cv=cv, scoring='accuracy')

print(f"5-Fold CV Accuracy scores: {scores}")
print(f"Mean CV Accuracy: {scores.mean():.4f}")
```

```python
import joblib

joblib.dump(clf, "lgbm.pkl")
print("✅ Model saved as: lgbm.pkl")

# # Load model
# # clf = joblib.load("lgbm_amazon.pkl")
```
