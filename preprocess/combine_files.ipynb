```python
MODEL = 'BERT'
PATH = 'cleaned'
COMPANIES = ['meta', 'amazon', 'apple', 'microsoft', 'google']
ROLES = ['da', 'ds', 'pm', 'swe']
```

```python
import pandas as pd

# Load CSV files
files = []
for comp in COMPANIES:
    for role in ROLES:
        if MODEL == 'BERT':
            files.append(f'cleaned_bert_{comp}_{role}.csv')
        else:
            files.append(f'cleaned_{comp}_{role}.csv')

print("Files to concat:", files)

# Read files and concat
path = f'{PATH}/{MODEL}'
df_list = [pd.read_csv(f'{path}/{fn}') for fn in files]
df = pd.concat(df_list, ignore_index=True)

# 3. Find all comp_* columns. Fill NaN in 0
comp_cols = [c for c in df.columns if c.startswith('comp_')]
df[comp_cols] = df[comp_cols].fillna(0).astype(int)

# Save file
out = f'{path}/FAAMG_{"bert" if MODEL=="BERT" else "all"}.csv'
df.to_csv(out, index=False)
print(f"Merging completed. Ouput file: {out}")
```
