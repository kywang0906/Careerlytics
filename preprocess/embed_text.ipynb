```python
PATH = 'cleaned/LGBM'
```

```python
import pandas as pd
from sentence_transformers import SentenceTransformer

df = pd.read_csv(f"{PATH}/FAAMG_all.csv")

# Initialize Sentence-BERT
model = SentenceTransformer("distiluse-base-multilingual-cased-v2")
text_columns = ["about", "cert_text", "course_text", "pub_text", "proj_text"]

for col in text_columns:
    # Fill NaN and convert df to list
    texts = df[col].fillna("").tolist()
    embeddings = model.encode(texts, show_progress_bar=True)
    # Convert to DataFrame and add prefix
    emb_df = pd.DataFrame(embeddings, index=df.index).add_prefix(f"{col}_emb_")
    # Merge back to df
    df = pd.concat([df, emb_df], axis=1)

columns_to_drop = [
    'about', 'cert_text', 'course_text', 'pub_text', 'proj_text'
]

df_final = df.drop(columns=columns_to_drop, errors='ignore')

df_final.to_csv(f"embedded/FAAMG_emb.csv", index=False)
print(f"SBERT embeddings completed. File saved to FAAMG_emb.csv")
```
